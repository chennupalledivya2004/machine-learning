```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt

jc_matrix = np.zeros((20, 20))
smc_matrix = np.zeros((20, 20))
cosine_matrix = np.zeros((20, 20))

for i in range(20):
    for j in range(i, 20):
        f11 = sum([1 for a, b in zip(df.iloc[i], df.iloc[j]) if a == b == 1])
        f01 = sum([1 for a, b in zip(df.iloc[i], df.iloc[j]) if a == 0 and b == 1])
        f10 = sum([1 for a, b in zip(df.iloc[i], df.iloc[j]) if a == 1 and b == 0])
        f00 = sum([1 for a, b in zip(df.iloc[i],df.iloc[j]) if a == b == 0])
        jc = f11 / (f01 + f10 + f11)
        smc = (f11 + f00) / (f00 + f01 + f10 + f11)
        jc_matrix[i, j] = jc
        jc_matrix[j, i] = jc
        smc_matrix[i, j] = smc
        smc_matrix[j, i] = smc

cosine_matrix = cosine_similarity(df)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.heatmap(jc_matrix, annot=True, cmap='coolwarm', ax=axes[0], square=True)
axes[0].set_title('Jaccard Coefficient')

sns.heatmap(smc_matrix, annot=True, cmap='coolwarm', ax=axes[1], square=True)
axes[1].set_title('Simple Matching Coefficient')

sns.heatmap(cosine_matrix, annot=True, cmap='coolwarm', ax=axes[2], square=True)
axes[2].set_title('Cosine Similarity')

plt.tight_layout()
plt.show()
```
