```python
import pandas as pd
import numpy as np

df=pd.read_excel('impurity.xlsx')
df
```

```python
class_counts = df['student'].value_counts()

total_instances = len(df)
prior_prob_yes = class_counts['yes'] / total_instances
prior_prob_no = class_counts['no'] / total_instances

# Print the results
print("Prior probability of buys_computer = yes:", prior_prob_yes)
print("Prior probability of buys_computer = no:", prior_prob_no)
```

```python
import pandas as pd
from sklearn.neighbors import KernelDensity
numerical_features = ['age']

for class_label in df['student'].unique():
    print(f"Class: {class_label}")
    for feature in numerical_features:

        data_subset = df[df['student'] == class_label][feature]
        data_subset = pd.to_numeric(data_subset, errors='coerce')  # Convert to numeric, coerce non-numeric to NaN
        data_subset = data_subset.dropna()  # Drop NaN values

        if not data_subset.empty:
            # Fit kernel density
            kde = KernelDensity(bandwidth=1.0, kernel='gaussian')
            kde.fit(data_subset.values.reshape(-1, 1))
            values_to_evaluate = [30, 35, 40]
            densities = np.exp(kde.score_samples(np.array(values_to_evaluate).reshape(-1, 1)))

            print(f"Feature: {feature}")
            print(f"Densities: {densities}")
            print(f"Any zero density: {any(d == 0 for d in densities)}")
            print()
        else:
            print(f"Feature: {feature} - No numeric values for this class")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(index=df['age'], columns=[df['income'], df['student'], df['buys_computer']])

# Chi-square test of independence
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)
alpha = 0.05
print("\nTest result:")
if p_value < alpha:
    print("Reject the null hypothesis. There is evidence of a significant association between the features.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between the features.")
```
